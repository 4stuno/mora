# Notebook de Raciocínio DL

Este notebook demonstra o uso do reasoner DL para executar inferências sobre a ontologia.


```python
import sys
import os
sys.path.append(os.path.dirname(os.path.dirname(os.path.abspath(''))))

from ontology.reasoner import DLReasoner
```

## 1. Carregar Ontologia e Inicializar Reasoner


```python
reasoner = DLReasoner(ontology_path="ontologia_mora.owl")
```

## 2. Classificação


```python
classification = reasoner.classify()
print(f"Classes com subclasses: {len(classification)}")
for cls, subclasses in list(classification.items())[:10]:
    print(f"\n{cls}:")
    for sub in subclasses[:5]:
        print(f"  - {sub}")
```
